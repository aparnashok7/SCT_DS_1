```python
import matplotlib.pyplot as plt

# Sample data
gender = ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']

# Create a bar chart
plt.bar(gender, [1]*len(gender))
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
ages = np.random.randint(18, 80, 100)  # 100 random ages between 18 and 80

# Create a histogram
plt.hist(ages, bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
gender = ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']

# Create a bar chart
sns.countplot(x=gender)
plt.title('Distribution of Genders')
plt.show()
```
